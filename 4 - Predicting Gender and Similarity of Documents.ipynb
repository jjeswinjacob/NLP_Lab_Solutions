```python
# TASK 2 SIMPLE TEXT CLASSIFICATION
def gender_features(word):
    return {'last_letter': word[-1]}
```

```python
gender_features('Obama')
```

```python
from nltk.corpus import names
#names.word()
print(len(names.words()))
```

```python
labeled_names = ([(name, 'male') for name in names.words('male.txt')] + [(name, 'female') for name in names.words('female.txt')])
labeled_names
```

```python
import random
random.shuffle(labeled_names)
```

```python
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]
```

```python
featuresets
```

```python
train_set, test_set = featuresets[:5000], featuresets[5000:]
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('David'))
```

```python
classifier.classify(gender_features('Yann Lecun'))
```

```python
classifier.classify(gender_features('Sharmila Banu'))
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
from nltk.corpus import inaugural
from sklearn.feature_extraction.text import CountVectorizer
obamaSpeech = inaugural.words(fileids = '2009-Obama.txt')
trumpSpeech = inaugural.words(fileids = '2017-Trump.txt')
obamaSpeech = " ".join(obamaSpeech)
trumpSpeech = " ".join(trumpSpeech)
documents = [str(obamaSpeech), str(trumpSpeech)]

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Create the Document Term Matrix
count_vectorizer = CountVectorizer(stop_words='english')
sparse_matrix = count_vectorizer.fit_transform(documents)

doc_term_matrix = sparse_matrix.todense()
df = pd.DataFrame(doc_term_matrix, columns=count_vectorizer.get_feature_names(), index=['obamaSpeech', 'trumpSpeech'])
print(df)

from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(sparse_matrix[0:1], sparse_matrix)
```

```python
# Finding Similarity with stemming

from nltk.corpus import inaugural
from sklearn.feature_extraction.text import CountVectorizer
obamaSpeech = inaugural.words(fileids = '2009-Obama.txt')
trumpSpeech = inaugural.words(fileids = '2017-Trump.txt')

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
obamaSpeech = [stemmer.stem(token) for token in obamaSpeech]
trumpSpeech = [stemmer.stem(token) for token in trumpSpeech]

obamaSpeech = " ".join(obamaSpeech)
trumpSpeech = " ".join(trumpSpeech)
documents = [str(obamaSpeech), str(trumpSpeech)]

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Create the Document Term Matrix
count_vectorizer = CountVectorizer(stop_words='english')
sparse_matrix = count_vectorizer.fit_transform(documents)

doc_term_matrix = sparse_matrix.todense()
df = pd.DataFrame(doc_term_matrix, columns=count_vectorizer.get_feature_names(), index=['obamaSpeech', 'trumpSpeech'])
print(df)

from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(sparse_matrix[0:1], sparse_matrix)
```

# Conclusion: We're able to achieve a similarity of 48.53% without stemming and a 59.084% with stemming.
# We're able to see that a little preprocessing helps us get more similarity. Hence, stemming is uselful in finding similarity between two documents and aso in understanding plagiarism
# Google search engine would use text preprocessing techniques of stemming and lemmatization to achieve a greater accuracy of search results. This helps in providing more number of results to the end user.
