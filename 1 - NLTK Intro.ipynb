# TASK 5: FREQUENCY DISTRIBUTION OF WORDS IN A TEXT

```python
text1 = 'Hi Hi Hi Lo Lo Ki'
fd = nltk.FreqDist(text1.split())
fd
```

# TASK 6. CONDITIONAL FREQUENCY DISTRIBUTION OF WORD IN A TEXT

```python
from nltk.probability import ConditionalFreqDist
cfd = ConditionalFreqDist((len(word), word) for word in text1.split())
cfd[2]
```

# HW 1. To Determine frequency Distribution and Conditional Frequency of any of the presidential inaugral speeches

```python
obamaSpeech = inaugural.words(fileids = '2009-Obama.txt')
fd = nltk.FreqDist(obamaSpeech)
print(fd['America'])
cfd = ConditionalFreqDist((len(word), word) for word in obamaSpeech)
cfd[14]
```

# Conclusion

```python
maxLen = 0
for i in range(0, 100):
    if len(cfd[i]) != 0 and i > maxLen:
        maxLen = i
print(maxLen)
cfd[maxLen]
```
