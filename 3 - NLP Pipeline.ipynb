```python
# TASK 1: BASIC TEXT PROCESSING PIPELINE
import nltk
texts = """When exercising, count backwards.
For example, if you are carrying out 20 sit ups, don’t count from 1 to 20, start at 20 and count backwards as you do them.
This is a psychological trick to make a large number of tough exercises seem more achievable and less daunting.
Start listening to your gut instinct. 
It’s always right. 
Scientists have studied the ‘gut instinct’ in depth and learned that the dopamine neurons in your body alert you when something or someone in your environment is ‘off’.
If you get a bad feeling about a situation, listen to it. It’s happening for a reason."""


sentences = nltk.sent_tokenize(texts)
print(sentences, '\n')
for sentence in sentences:
    words = nltk.word_tokenize(sentence)
    print("Words in ", sentence, words, '\n')
    tagged = nltk.pos_tag(words)
    print(tagged, '\n')
```

```python
#TASK 3: SCRAPING DATA FROM WEB
from urllib import request
url = "https://www.gutenberg.org/files/2554/2554-0.txt"
response = request.urlopen(url)
raw = response.read().decode('utf8')
type(raw)
print(raw[:50])
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(raw)
type(tokens)
# print(tokens)

text = nltk.Text(tokens)
print(type(text))
# print(text)

words = [w.lower() for w in tokens]
vocab = sorted(set(words))
# type(vocab)
print(vocab)

# HTML => ASCII => TOKEN => TEXT => VOCAB
# HW: TOKEN => TEXT => VOCAB
```
